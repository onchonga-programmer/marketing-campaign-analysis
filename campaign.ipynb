```python
#rename the columns
df.rename(columns={
    'Unique Link Clicks (ULC)': 'unique_link_clicks',
    'Cost Per Click (CPC)': 'cost_per_click',
    'Click-Through Rate (CTR in %)':'ctr',
    'Unique Click-Through Rate (Unique CTR in %)': 'unique_ctr',
    'Amount Spent in INR':'amount',
    'Cost per Result (CPR)': 'cost_per_result',
},inplace=True)
```

```python
# remove unwanted characters from numeric columns
for col in ['amount', 'cost_per_click', 'cost_per_result']:
    df[col] = df[col].replace('[\$,]', '', regex=True).astype(float)

for col in ['ctr', 'unique_ctr']:
    df[col] = df[col].astype(float)
df.head()
```

```python
#fix misalligned rows
df_clean =df.dropna(thresh=10)
#print head of the cleaned dataframe
df_clean.head()

```

# data visualisation
## Feature Description

| **Feature** | **Description** |
|-------------|------------------|
| `campaign_id`, `campaign_name` | Identifier and name of each campaign |
| `audience`, `age`, `geography` | Target audience characteristics |
| `reach`, `impressions`, `frequency` | Ad performance metrics:<br> **Reach**  number of unique users who saw the ad <br> **Impressions**  total number of times the ad was shown <br> **Frequency**  average number of times the ad was shown per user |
| `clicks`, `unique_clicks`, `unique_link_clicks` | Engagement metrics:<br> **Clicks**  total clicks on the ad <br> **Unique Clicks**  number of individual users who clicked <br> **Unique Link Clicks**  distinct users who clicked the main link |
| `ctr`, `unique_ctr` | Click-through rates:<br> **CTR**  (clicks  impressions)  100 <br> **Unique CTR**  (unique clicks  impressions)  100 |
| `amount_spent` | Total budget spent on a specific campaign segment |
| `cost_per_click`, `cost_per_result`, `calculated_cpc` | Cost efficiency metrics:<br> **Cost Per Click (CPC)**  average cost for each click <br> **Cost Per Result**  cost per outcome (e.g., signup) <br> **Calculated CPC**  additional verification CPC calculation |
| `calculated_ctr` | An alternate computation of the click-through rate for validation |

```python
# group campaign perfomance by campaign name
campaign_summary = df_clean.groupby('campaign ID').agg({
    'amount': 'sum',
    'Clicks': 'sum',
    'Impressions': 'sum',
    'Reach': 'sum',
    'ctr': 'mean',
    'cost_per_click': 'mean',
    'cost_per_result': 'mean',
}).sort_values(by='amount', ascending=False).reset_index()
campaign_summary
```

```python
#total amount vs clicks
plt.figure(figsize=(12,6))
sns.barplot(data=campaign_summary, x='campaign ID', y='amount', color='skyblue', label='Amount Spent')
sns.lineplot(data=campaign_summary, x='campaign ID', y='Clicks', color='red', marker='o', label='Clicks')

plt.title('Total Amount Spent vs Clicks per Campaign')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#scatter plot of cost vs click rate
plt.figure(figsize=(12, 6))
plt.figure(figsize=(10,6))
sns.scatterplot(data=campaign_summary,x='cost_per_click',y='ctr' ,hue='campaign ID',palette='tab10',s=100)
plt.title('cost per click vs click rate')
plt.show()
```

```python
if 'Unique Clicks' in df.columns:
    clicks_df = df.groupby('campaign ID')['Unique Clicks'].sum().reset_index()
    merged = campaign_summary.merge(clicks_df, on='campaign ID', how='left')

    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=merged, x='Impressions', y='Unique Clicks', hue='campaign ID', palette='tab10', s=100)

    plt.title('Impressions vs Unique Clicks')
    plt.xlabel('Impressions')
    plt.ylabel('Unique Clicks')
    plt.tight_layout()
    plt.show()
else:
    print("Column 'Unique Clicks' not found in the dataset.")
```

```python
#create barplot of impressions vs unique clicks
plt.figure(figsize=(12, 6))
sns.barplot(data=merged, x='Impressions', y='Unique Clicks', hue='campaign ID', palette='tab10')
plt.title('Impressions vs Unique Clicks per Campaign')
plt.xlabel('Impressions')
plt.ylabel('Unique Clicks')
plt.xticks(rotation=45) 
plt.tight_layout()
plt.show()
```

```python
# Click-Through Rate (CTR)
df['calculated_ctr'] = df['Clicks'] / df['Impressions']

# Cost Efficiency
df['cost_efficiency'] = df['cost_per_result'] / df['Clicks']
```

```python
low_performers = df[
    (df['cost_per_result'] > df['cost_per_result'].median()) &
    (df['ctr'] < df['ctr'].median())
]

print(low_performers[['campaign ID', 'cost_per_result', 'ctr']])
```

```python
# Group by Age and summarize relevant metrics
age_group_summary = df.groupby('Age').agg({
    'Reach': 'sum',
    'Impressions': 'sum',
    'Clicks': 'sum',
    'ctr': 'mean',
    'cost_per_click': 'mean',
    'cost_per_result': 'mean',
    "amount": 'sum'
}).reset_index()

print(age_group_summary)
```

```python
sns.barplot(data=age_group_summary, x='Age', y='ctr')
plt.title('Click Through Rate by Age Group')
plt.xticks(rotation=45)
plt.show()
```

```python
# Set style
sns.set(style='whitegrid')

# Create a barplot for CTR and CPC by age
fig, ax1 = plt.subplots(figsize=(10, 5))

# Barplot for CTR
sns.barplot(x='Age', y='ctr', data=age_group_summary, color='skyblue', label='CTR', ax=ax1)
ax1.set_ylabel('Click Through Rate (CTR)')
ax1.set_title('CTR and CPC by Age Group')

# Create a second y-axis for CPC
ax2 = ax1.twinx()
sns.lineplot(x='Age', y='cost_per_click', data=age_group_summary, color='red', marker='o', label='CPC', ax=ax2)
ax2.set_ylabel('Cost Per Click (CPC)', color='red')
ax2.tick_params(axis='y', labelcolor='red')

plt.tight_layout()
plt.show()
```
